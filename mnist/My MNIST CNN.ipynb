```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
PATH="data/"
```

```python
#important for monitoring metrics in tensorboard
tf.logging.set_verbosity(tf.logging.INFO)
```

Load the data:

```python
mnist_train_df = pd.read_csv(f'{PATH}train.csv', sep = ",", header = [0])
mnist_train_df = mnist_train_df.reindex(np.random.permutation(mnist_train_df.index))
mnist_train_df.describe()
```

```python
mnist_test_df  = pd.read_csv(f'{PATH}test.csv', sep = ",", header = [0])
mnist_test_df.describe()
```

```python
def image_preprocess(df):
    df = df/255.0
    df = df.values.tolist()
    arr = np.asarray(df, dtype = np.float32)
    return arr
```

Split the data into train and validation:

```python
SPLIT = 37800
```

```python
train_images = image_preprocess(mnist_train_df.iloc[:SPLIT,1:])
train_labels = np.asarray(mnist_train_df.iloc[:SPLIT,0], dtype = np.int32)

valid_images = image_preprocess(mnist_train_df.iloc[SPLIT:,1:])
valid_labels = np.asarray(mnist_train_df.iloc[SPLIT:,0], dtype = np.int32)

test_images = image_preprocess(mnist_test_df)

print(train_images.shape)
print(train_labels.shape)

print(valid_images.shape)
print(valid_labels.shape)

print(test_images.shape)
train_labels
```

```python
BATCH_SIZE = 64
NUM_EPOCHS = 5
```

Define input functions:

```python
train_input_fn = tf.estimator.inputs.numpy_input_fn(x={'x' : train_images},
                                                    y=train_labels,
                                                    batch_size = BATCH_SIZE,
                                                    num_epochs=1,
                                                    shuffle=True)
```

```python
valid_input_fn = tf.estimator.inputs.numpy_input_fn(x={'x' : valid_images},
                                                   y=valid_labels,
                                                   batch_size = BATCH_SIZE,
                                                   num_epochs=1,
                                                   shuffle=False)
```

```python
test_input_fn = tf.estimator.inputs.numpy_input_fn(x={'x' : test_images},
                                                  batch_size = BATCH_SIZE,
                                                  num_epochs=1,
                                                  shuffle=False)
```

Define model function:

```python
def conv_model(features, labels, mode):
    XX = tf.reshape(features['x'], [-1,28,28,1])
    Y1 = tf.layers.conv2d(XX, filters=6, kernel_size=[6,6], padding='same', activation=tf.nn.leaky_relu)
    Y2 = tf.layers.conv2d(Y1, filters=12, kernel_size=[5,5], strides=(2,2), padding='same', activation=tf.nn.leaky_relu)
    Y3 = tf.layers.conv2d(Y2, filters=24, kernel_size=[4,4], strides=(2,2), padding='same', activation=tf.nn.leaky_relu)
    Y4 = tf.layers.flatten(Y3)
    Y5 = tf.layers.dense(Y4, units = 200, activation=tf.nn.leaky_relu)
    Y6 = tf.layers.dropout(Y5, rate = 0.1, training=(mode == tf.estimator.ModeKeys.TRAIN))
    logits = tf.layers.dense(Y6,units = 10)
    
    predictions = {
                    "classes": tf.argmax(input=logits, axis=1),
                    "probabilities": tf.nn.softmax(logits)
    }
    
    #define predict method logic
    
    if mode == tf.estimator.ModeKeys.PREDICT:
        return tf.estimator.EstimatorSpec(mode = mode, predictions=predictions)
    
    loss = tf.losses.sparse_softmax_cross_entropy(labels = labels, logits = logits)
    accuracy = tf.metrics.accuracy(labels = labels, predictions = predictions["classes"], name = "acc_op")
    
    
    #define train method logic
    
    if mode == tf.estimator.ModeKeys.TRAIN:
        optimizer = tf.train.AdamOptimizer()
        train_op  = optimizer.minimize(
            loss = loss,
            global_step=tf.train.get_global_step())
        tf.identity(accuracy[1], name='train_accuracy')
        tf.summary.scalar('train_accuracy', accuracy[1])
        eval_metric_ops = {"train_accuracy":accuracy}
        return tf.estimator.EstimatorSpec(mode = mode, loss = loss, train_op = train_op, eval_metric_ops=eval_metric_ops)
    
    
    
    #define evaluate method logic
    tf.identity(accuracy[1], name='val_accuracy')
    tf.summary.scalar('val_accuracy', accuracy[1])
    eval_metric_ops = {"val_accuracy":accuracy}
    return tf.estimator.EstimatorSpec(mode = mode, loss = loss, eval_metric_ops=eval_metric_ops)
```

Train the classifier:

```python
OUT_DIR = "./tf_files"
```

```python
my_cnn_classifier = tf.estimator.Estimator(model_fn=conv_model, model_dir=OUT_DIR)
file_writer = tf.summary.FileWriter(OUT_DIR)

def train_and_eval(estimator, num_epochs = NUM_EPOCHS):
    for n in range(num_epochs):
        estimator.train(input_fn = train_input_fn)
        estimator.evaluate(input_fn = valid_input_fn)
        
train_and_eval(my_cnn_classifier)
```

Generate submission results:

```python
generator = my_cnn_classifier.predict(input_fn=test_input_fn)
predictions = [next(generator) for i in range(len(test_images))]
classes = [predictions[i]["classes"] for i in range(len(predictions))]

def make_submission_file(classes, filename):
    submission = pd.DataFrame()
    submission["ImageId"] = range(1, 28001)
    submission["Label"] = classes
    submission.set_index("ImageId", inplace=True)
    submission.to_csv(filename)
    
make_submission_file(classes, "mtolstok_mnist_subm.csv")
```
